## 1. Џупитер радна свеска

За разлику од других окружења у којима се могу извршавати Пајтон програми, Џупитер радна свеска се састоји из низа ћелија, а свака ћелија може да садржи текст, математички израз или низ Пајтон наредби. На тај начин цео процес обраде података остаје документован у облику приче, може се показати сарадницима, може се послати електронском поштом, или се може лако модификовати ради понављања сличног процеса.

Низ икона при врху стране ти омогућује да бараташ ћелијама Џупитер радне свеске на разне начине.

Дугме **Run** смо већ користили. Погледајмо сада чему служе још четири дугмета:

![Četiri dugmeta](slike/JupyterMenu4Buttons.jpg)

- Клик на дугме које личи на дискету (прва слева) ће снимити радну свеску.
- Клик на дугме са знаком <b>+</b> ће додати нову ћелију испод активне ћелије. (Ћелију активираш тако што кликнеш на њу; активна ћелија има <font color=green>***зелени***</font> или <font color=blue>***плави***</font> оквир.)
- Стрелице на горе и на доле померају радну ћелију за једно место горе или доле.

На пример, читаш овај текст и одједном осетиш снажну жељу да израчунаш вредност израза $1 + \frac 12 + \frac 13 + \frac 14 + \frac 15 + \frac 16 + \frac 17$. Тада треба да кликнеш на ову ћелију (коју управо читаш) и онда на дугме са знаком **+**. У новој ћелији која ће се појавити испод можеш да унесеш израз, а Пајтон ће га израчунати.

У осам ћелија које следе наведени су стихови познате енглеске дечије бројалице. Стихови су испретурани. Користећи дугмиће са стрелицама преуреди ћелије тако да добијеш текст бројалице.

*Once I caught a fish alive,*

*Then I let it go again.*

*Which finger did it bite?*

*Why did you let it go?*

*Because it bit my finger so.*

*This little finger on my right.*

*One, two, three, four, five,*

*Six, seven, eight, nine, ten,*

Следећа три дугмета:

![Još tri dugmeta](slike/JupyterMenuCutCopyPaste.jpg)

омогућују да се са ћелијама обављају стандардне операције које се на енглеском зову *cut* (исеци, дугме "маказе"), *copy* (ископирај, дугме "два листа папира") и *paste* (залепи, дугме "лист папира који се додаје на радну подлогу").

Ако желиш **да преместиш неку ћелију:**
1. кликни на ту ћелију,
2. кликни на дугме *cut* (дугме "маказе"),
3. кликни на неко друго место у Џупитер радној свесци, и
4. кликни на дугме *paste* (дугме "лист папира који се додаје на радну подлогу").

Ова операција се кратко зове *cut/paste*.

Ако желиш **да ископираш неку ћелију:**
1. кликни на ту ћелију,
2. кликни на дугме *copy* (дугме "два листа папира")
3. кликни на неко друго место у Џупитер радној свесци, и
4. кликни на дугме *paste* (дугме "лист папира који се додаје на радну подлогу").

Ова операција се кратко зове *copy/paste*.

Џупитер ћелија може да садржи и текст. Текст се уноси у посебном облику који се зове *Markdown*.

Додај ћелију испод и унесе неки текст, али тако да сваки од следећих елемената буде заступљен:
- наслов и поднаслов,
- **подебљан текст**,
- *италик*,
- навођење са туфницама,
- набрајање.

## 3. Шта је кернел и зашто би неко пожелео да га прекине или рестартује?

Када комуницирамо са Џупитером, то шематски изгледа овако:

![Komunikacija sa Jupyterom](slike/notebook_components.png)

(слика преузета са jupyter.org)

**Кернел** (енгл. "језгро") је "машина" која извршава команде и памти податке. То је главни погон радне свеске.

Свака нова радна свеска коју отворимо:
- отвори нови таб у прегледачу (browser), и
- покрене своју копију кернела.

Према томе, колико радних свески отворимо, толико копија кернела ће бити покренуто.

**Затварањем таба у прегледачу се НЕ затвара кернел који је због те радне свеске отворен!**

Да би се затворио таб **И** кернел, треба одабрати

File -> Close and Halt

Рестартовање кернела:

Kernel -> Restart

Прекидање кернела (када се упетљамо у бесконачну петљу):

Kernel -> Interrupt
